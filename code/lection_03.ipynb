## Clustering in Scipy

```python
from scipy.cluster import hierarchy
from matplotlib import pyplot as plt
```

```python
Z = hierarchy.linkage(X, method = "ward")
```

```python
hierarchy.dendrogram(Z)
plt.show()
```
